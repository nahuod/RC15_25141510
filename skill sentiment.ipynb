```python
import pandas as pd
!pip install vaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer


print("\n=== CSV File Head ===")

print("=====================\n")


df = pd.read_excel("/content/flickr总.xlsx")


print("DataFrame columns:", df.columns)


analyzer = SentimentIntensityAnalyzer()

examples = [
    "I love this! It's amazing and makes me so happy.",
    "This is the worst. I hate it and I'm disappointed.",
    "I'm not sure how I feel about this...",
    "LOL that was hilarious! :D",
    "The product is okay, but not great."
]

print("=== example ===")
for text in examples:
    scores = analyzer.polarity_scores(text)
    print(text)
    print("  ->", scores)


# -------------------------------
sentiment_results = []
print("\n=== example ===")


for review_text in df["description"].fillna(""):
    scores = analyzer.polarity_scores(str(review_text))
    sentiment_results.append(scores)

    print(review_text)
    print("  ->", scores)


# -------------------------------
# -------------------------------
df["sent_neg"] = [s["neg"] for s in sentiment_results]
df["sent_neu"] = [s["neu"] for s in sentiment_results]
df["sent_pos"] = [s["pos"] for s in sentiment_results]
df["sent_compound"] = [s["compound"] for s in sentiment_results]

def label_sentiment(c):
    if c >= 0.05:
        return "positive"
    elif c <= -0.05:
        return "negative"
    else:
        return "neutral"

df["sent_label"] = df["sent_compound"].apply(label_sentiment)


# -------------------------------
output_path = "/content/sentiment.csv"
df.to_csv(output_path, index=False)

print(f"\n save：{output_path}")
```
